## Age correcting

```python
def r(s):
    s = s.replace('ساله', '')
    s = s.replace(' ', '')
    s = s.replace('-', '000')
    s = s.replace('85وبيشتر' , '85')
    s = s.replace('85وبیشتر' , '85')
    s = s.replace('سننامشخص' , '99')
    s = s.replace('سنیننامشخص' , '99')
    s = s.replace('نامشخص' , '99')
    s = s.replace('نامشخص' , '99')
    s = s.replace('كمترازيك' , '0')

    return(s)
d.age = d.age.apply(r)
d = d[d.age.str.len()<=2]
d.age = d.age.apply(lambda x:int(x))
# d.groupby('age').count()

```
